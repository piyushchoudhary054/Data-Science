```python

# In this dataset, there are two numerical columns: age and fever.
# 1. Count
    # Number of non-missing values.
    # Age: 100 values are available.
    # Fever: Only 90 values are available, meaning 10 values are missing.
# 2. Mean (Average)
    # Sum of all values divided by the number of observations.
    # Average age = 44.22 years
    # Average fever = 100.84°F
# 3. Standard Deviation (Std)
    # Measures how spread out the values are from the mean.
    # Age std = 24.88 → Ages vary widely.
    # Fever std = 2.05 → Fever temperatures are relatively close to the average.
# 4. Minimum (Min)
    # Smallest value in the column.
    # Youngest age = 5 years
    # Lowest fever = 98°F
# 5. 25% (First Quartile, Q1)
    # 25% of observations are below this value.
    # 25% of people are 20 years or younger.
    # 25% of fever readings are 99°F or lower.
# 6. 50% (Median)
    # Middle value when data is sorted.
    # Median age = 45 years
    # Median fever = 101°F
# 7. 75% (Third Quartile, Q3)
    # 75% of observations are below this value.
    # 75% of people are 66.5 years or younger.
    # 75% of fever readings are 102.75°F or lower.
# 8. Maximum (Max)
    # Largest value in the column.
    # Oldest age = 84 years
    # Highest fever = 104°F
```
