```python
#  Encoding
#     ---- > Transform your categorical data into Numerical data .
# (1). Label ----- > It will encode data 1-1 columns .
# (2). Ordinal--- > It will encode data more than 1 columns .
# (3). OneHot ---- > It will create new column based on sub-categories but it has only values not column name.
# (4). get_dummies --- > It will create new column based on sub-categories but it shows column name with values


```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\PIYUSH\OneDrive\Desktop\Data-Science\Datasets\covid_toy - covid_toy.csv')
```

```python
df.head(3)
```

```python
df = df.drop(columns = ['age','fever'])
```

```python
df.head(3)
```

```python
# drop='first' → Har category group ki pehli category hata deta hai, taaki duplicate information na rahe.
# sparse_output=False → Output normal NumPy array mein deta hai, sparse matrix nahi.
# dtype=np.float32 → Output values ka data type float32 rakhta hai (kam memory use hoti hai).
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.float32)
```

```python
# gender --->2 --->1
# city --->4 ----->3
# has_covid --->2 --->1
# cought --->2 --->1
```

```python
ohe.fit_transform(df[['gender','cough','city','has_covid']])
```
