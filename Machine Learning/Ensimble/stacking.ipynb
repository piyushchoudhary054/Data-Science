```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('iris')
```

```python
df.head(3)
```

```python
# in cross validation we don't do train test split because it will be done by cross validation itself. So we will not do train test split here.
# but in ensemble learning we have to do train test split because we will train the model on training data and then we will test it on testing data.

```

```python
x = df.drop(columns='species')
y = df['species']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
y_encoder = lb.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
base_learner = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svc', SVC(probability=True,kernel='rbf',random_state=42)),
    ('lr',LogisticRegression(max_iter=1000))
]
```

```python
meta_learner = LogisticRegression(max_iter=1000)
```

```python
stacking_clf = StackingClassifier(
    estimators=base_learner,
    final_estimator=meta_learner,
    cv=5
)
```

```python
stacking_clf.fit(x_train,y_train)
```

```python
y_pred = stacking_clf.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```
