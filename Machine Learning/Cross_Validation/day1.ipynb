```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=['deck','alive','embark_town','adult_male','who','class'])
```

```python
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
col = df.select_dtypes(include='object').columns
col
```

```python
for i in col:
    df[i] = lb.fit_transform(df[i])
```

```python
df = df.astype('int')
```

```python
df.info()
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns='survived')
y = df['survived']
```

```python
# x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# now we will scale this as we are usig standard scaler
from sklearn.preprocessing import StandardScaler
```

```python
sc  = StandardScaler()
```

```python
x_scale = sc.fit_transform(x)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='linear')
# svm = SVC(kernel='rbf')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
svm_scores = cross_val_score(svm,x_scale,y, cv=5, scoring = 'accuracy')
knn_scores = cross_val_score(knn,x_scale,y, cv=5, scoring = 'accuracy')
```

```python
print(svm_scores.mean())
print(knn_scores.mean())
```
