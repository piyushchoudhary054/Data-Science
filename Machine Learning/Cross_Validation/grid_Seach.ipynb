```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
df['species'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns='species')
y = df['species']
```

```python
# cross validation we don't do train test split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
# if k = 15 accuracy is 1 
# knn = KNeighborsClassifier(n_neighbors=15)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
model_svm = SVC(C=30,kernel='linear',gamma='auto')
# model_svm = SVC(C=30,kernel='rbf',gamma='auto')
```

```python
model_svm.fit(x_train,y_train)
```

```python
model_svm.score(x_test,y_test)
```

```python
# how to use grid search cv
# cv -> cross validation 
# cv means jo data set hai lets say 100 % hai let say agar apan ne cv 5 raka to 
# uske 5 parts bnyega and har ek part se test krega bakio ko chod k
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC(gamma='auto')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svc),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear'],
},cv=5,return_train_score=False)
```

```python
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
# Data --> mode ---> cv -->grid
```

```python
classifier_r.fit(x,y)
```
