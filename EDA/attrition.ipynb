
Employee attrition analysis is a type of
behavioural analysis where we study the
behaviour and characteristics of the employees
who left the organization and compare their
characteristics with the current employees to
find the employees who may leave the organization soon.


A high rate of attrition of employees can be expensive for any company in terms of recruitment and
training costs, loss of productivity and morale reduction of employees. By identifying the causes of
attrition, a company can take measures to reduce the attrition of employees and maintain precious
employees.

```python
categorical_Col = df.select_dtypes(include='object').columns
cat = categorical_Col
numerical_Col = df.select_dtypes(exclude='object').columns
num = numerical_Col
```

```python
categorical = df[categorical_Col]
```

```python
categorical.head()
```

```python
categorical['Attrition'] = categorical['Attrition'].map({'Yes': 1, 'No': 0})
categorical['OverTime'] = categorical['OverTime'].map({'Yes': 1, 'No': 0})
categorical['BusinessTravel'] = categorical['BusinessTravel'].map({'Non-Travel': 0, 'Travel_Rarely': 1, 'Travel_Frequently': 2})
categorical['Department'] = categorical['Department'].map({'Sales': 0, 'Research & Development': 1, 'Human Resources': 2})
categorical['EducationField'] = categorical['EducationField'].map({'Life Sciences': 0, 'Medical': 1, 'Marketing': 2, 'Technical Degree': 3, 'Human Resources': 4, 'Other': 5})
categorical['Gender'] = categorical['Gender'].map({'Male': 0, 'Female': 1})
categorical['JobRole'] = categorical['JobRole'].map({'Sales Executive': 0, 'Research Scientist': 1, 'Laboratory Technician': 2, 'Manufacturing Director': 3, 'Healthcare Representative': 4, 'Manager': 5, 'Sales Representative': 6, 'Research Director': 7, 'Human Resources': 8})    
categorical['MaritalStatus'] = categorical['MaritalStatus'].map({'Single': 0, 'Married': 1, 'Divorced': 2})
categorical['Over18'] = categorical['Over18'].map({'Y': 1, 'N': 0})
```

```python
numerical = df[numerical_Col]
numerical.head()
```

```python
numerical = numerical.rename(columns={'EmployeeNumber': 'Employee_Number', 'StandardHours': 'Standard_Hours', 'EmployeeCount': 'Employee_Count', 'DailyRate': 'Daily_Rate', 'HourlyRate': 'Hourly_Rate', 'MonthlyRate': 'Monthly_Rate'})
```

```python
new_df = pd.concat([categorical, numerical], axis=1)
new_df.head()
```
