```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.isnull().sum()
```

# Bivariate Analysis

```python
# scatter plot(numerical vs numerical columns)
sns.scatterplot(x= df['total_bill'], y=df['tip'])
```

```python
# to get the relation between total bill  and tip with respect to the gender of the customer
sns.scatterplot(x= df['total_bill'], y=df['tip'], hue=df['sex'])
```

```python
sns.scatterplot(x= df['total_bill'], y=df['tip'], hue=df['sex'], style=df['smoker'])
```

```python
# heatmap  - categorical vs categorical columns
p = pd.crosstab(df['day'] , df['time'])
p

```

```python
sns.heatmap(p)
```
